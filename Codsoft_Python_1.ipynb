# **Task 1: To do list**

```python
class ToDoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, task):
        self.tasks.append(task)

    def view_tasks(self):
        if not self.tasks:
            print("Your to-do list is empty.")
        else:
            print("To-Do List:")
            for i, task in enumerate(self.tasks, start=1):
                print(f"{i}. {task['task']} - {'Done' if task['done'] else 'Not Done'}")

    def mark_task_complete(self, task_index):
        if 0 < task_index <= len(self.tasks):
            self.tasks[task_index - 1]['done'] = True
            print(f"Task {task_index} marked as complete.")
        else:
            print("Invalid task index.")

    def delete_task(self, task_index):
        if 0 < task_index <= len(self.tasks):
            del self.tasks[task_index - 1]
            print(f"Task {task_index} deleted.")
        else:
            print("Invalid task index.")


def main():
    to_do_list = ToDoList()

    while True:
        print("\nOptions:")
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Mark Task as Complete")
        print("4. Delete Task")
        print("5. Quit")

        choice = input("Enter your choice: ")

        if choice == "1":
            task = input("Enter the task: ")
            to_do_list.add_task({"task": task, "done": False})
        elif choice == "2":
            to_do_list.view_tasks()
        elif choice == "3":
            task_index = int(input("Enter the task index to mark as complete: "))
            to_do_list.mark_task_complete(task_index)
        elif choice == "4":
            task_index = int(input("Enter the task index to delete: "))
            to_do_list.delete_task(task_index)
        elif choice == "5":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please choose a valid option.")


if __name__ == "__main__":
    main()
```
