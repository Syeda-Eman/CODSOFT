# **Task 2: Calculator**

```python
# Define functions for basic operations
def add(x, y): return x + y
def subtract(x, y): return x - y
def multiply(x, y): return x * y
def divide(x, y): return x / y if y != 0 else "Cannot divide by zero"

# Dictionary to map user input to operations
operations = {
    'add': add,
    'subtract': subtract,
    'multiply': multiply,
    'divide': divide
}

# Main calculator loop
while True:
    print("Options: add, subtract, multiply, divide, or quit")
    user_input = input(": ").lower()

    if user_input == 'quit':
        break

    if user_input in operations:
        num1, num2 = map(float, input("Enter two numbers separated by space: ").split())
        result = operations[user_input](num1, num2)
        print("Result:", result)
    else:
        print("Invalid input. Please enter a valid operation.")
```
